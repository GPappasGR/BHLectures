```python
x,y = var('x y')
```

```python
r = 1/2*ln(x^2 + y^2)
```

```python
a = 0.9
f = r^2 -2*r + a^2*y^2/(x^2+y^2)
f
```

```python
ergo = region_plot(f <= 0, (x,-8, 8), (y, -8, 8), incol='lightgray', bordercol='grey')
```

```python
Rp = exp(1 + sqrt(1-a^2))
Rm = exp(1 - sqrt(1-a^2))
Hp = circle((0,0), Rp, color='black', thickness=2)
Hm = circle((0,0), Rm, color='green', thickness=2)
R0 = circle((0,0), 1, color='red', linestyle='dotted', thickness=3)
```

```python
show(ergo+Hp+Hm+R0)
```
