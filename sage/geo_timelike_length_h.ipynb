```python
f(x) = (sqrt(1-x^2) + asin(x)/x)/2
f(x)
```

```python
taylor(f(x), x, 0, 5)
```

```python
g = plot(f(x), (x, -1, 1), thickness=2, gridlines='major',
         axes_labels=[r'$h$', r'$L_{(p,q)}(\mathcal{C}_h)/T$'],
         fontsize=14)
g
```
